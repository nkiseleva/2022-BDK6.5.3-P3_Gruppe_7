```python
import urllib.request
import pandas as pd
#pandas importieren
```

```python
dataset = pd.read_csv("detailedrecords.csv", low_memory=False)

#pull the file
```

```python
dataset['Publication date (standardised)'].value_counts()
#check how many works were written in what year
```

```python
def convert_to_century(year):
    return str(year)[:2]+"00"

convert_to_century(1795)
#write a function to turn a year into a century
```

```python
dataset["Century"] = dataset["Publication date (standardised)"].apply(convert_to_century)
#apply the new function to the Publication dates and save it in the new colomn
```

```python
dataset
#see the new saved dataset with the colomn of centuries
```

```python
dataset['Century'].value_counts()
#check how many works were written in what century
```
